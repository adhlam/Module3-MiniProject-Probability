**Evaluating the Conditional Probability**


Bayes' Theorem is given by:

$$
P(AB) = \frac{P(AB)}{P(B)}=\frac{P(A).P(B|A)}{P(B)}
$$

where,

P(AB) = The probability of A given B

P(B) = The probability of B occuring

P(A) = The probability of A occuring

P(AB) = The probabilitiy of A and B occuring

P(BA) = The probability of B given A


$$
P(SpadeBlack) = \frac{P(SpadeBlack)}{P(Black)}=\frac{P(Spade).P(Black|Spade)}{P(Black)}
$$

```python
a = {(13/52)/(26/52)}

sum(a)
```

```python
b = {(13/52*1)/(26/52)}

sum(b)
```

Here, a = b = 0.5, therefore Byes' theorem is verified.

**Emperical Probability of:**

P(Diamonds) = 2/10 = 1/5

Its interesting to see the emperical probabilities are reaching towards theoretical probabilities even with such sample size. I pretty sure that within the first million sample size, the emperical probabilities will be virtually same as the theoretical probability.

```python
PseudoChoices =  choices(['Diamonds','Clubs','Hearts','Spades'],k=1000000)

countSpades = 0
countClubs = 0
countDiamonds = 0
countHearts = 0

for i in PseudoChoices:
  if i =='Diamonds':
    countSpades += 1
  elif i == 'Clubs':
    countClubs += 1
  elif i == 'Hearts':
    countDiamonds += 1   
  elif i == 'Spades':
    countHearts += 1

estimateProbs = [countSpades/len(PseudoChoices),countClubs/len(PseudoChoices),countDiamonds/len(PseudoChoices),countHearts/len(PseudoChoices)]
     
```
